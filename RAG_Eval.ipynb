```python
from parser import parse_gu_knowledge_base
from DB import VectorDB
from SGR import LLMClient, GenerationModule
from RAG import RAGSystem

# 1. Модуль отвечает за расчет эмбеддингов и поиск (Retrieval).
db = VectorDB(model_name='ai-forever/FRIDA')

# 2. Инициализация клиента для взаимодействия с LLM (GigaChat)
client = LLMClient()

# 3. Модуль генерации по методологии SGR (Social Generation Rule)
gen = GenerationModule(client)

# 4. Сборка RAG-системы 
rag = RAGSystem(db, gen)
```

```python
# 5. Наполнение базы знаний
rag.prepare_knowledge_base(parse_gu_knowledge_base)
```

```python
from evaluation import run_evaluation
```

```python
# 6. Оценка качества (Evaluation)
# Запуск автоматизированного тестирования на наборе данных tiny_eval.jsonl.
run_evaluation(rag, "tiny_eval.jsonl")
```
